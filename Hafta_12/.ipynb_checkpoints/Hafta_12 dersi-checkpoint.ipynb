```python
import pandas as pd
import numpy as np
from math import sqrt

# Visualization
from pylab import *
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import seaborn as sns

# Statistics
from statistics import median
from scipy import signal
import scipy.stats as stats
from scipy.stats import sem, binom, lognorm, poisson, bernoulli, spearmanr
from scipy.fftpack import fft, fftshift
```

```python
# PMF : Probability Mass Function (Olasılık Kütle Fonksiyonu)
# Olası değerlerin sayısı belirlidir.
p = 0.5 # Yazı-tura gelme ihtimali
n = 100

x = np.arange(binom.ppf(0.01, n, p), binom.ppf(0.99, n, p))
fig, ax = plt.subplots(1,1,figsize=(17,5))
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='Binomial PMF')
ax.vlines(x, 0, binom.pmf(x,n,p), colors='b', lw=5, alpha=0.5)
ax.legend(loc='best', frameon=False, fontsize='xx-large')
plt.title('Binom Dağılımı PMF (n=100, p=0.5)')
plt.show()
```

### PDF : Probabiliy Density Function (Olasılık Yoğunluk Fonksiyonu)
### Olası değerlerin sayısı sonsuzdur.
PDF Normal Dağılım


```python
# Plot Normal Distribution
mu = 0,
variance = 1
sigma = sqrt(variance)

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
Y = stats.norm.pdf(x, mu, sigma)
plt.figure(figsize=(16,5))
plt.plot(x, Y, label='Normal Dağılım')
plt.legend(fontsize='xx-large')
plt.show()
```

```python
# CDF (Cumulative Distribution Function) Kümülatif Dağılım Fonksiyonu
Y = Y / (0.01 * Y).sum()
plt.figure(figsize=(16,5))
plt.plot(x, np.cumsum(Y * 0.01), label='Cumulative Distribution Function(CDF)')
plt.legend(fontsize='xx-large')
plt.show()
```

```python
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
Y = stats.norm.pdf(x, mu, sigma)
plt.figure(figsize=(16,5))
plt.plot(x, Y, label='Probabiliy Density Function')
plt.legend(fontsize='xx-large')
Y = Y / (0.01 * Y).sum()
plt.plot(x, np.cumsum(Y * 0.01), label='Cumulative Distribution Function(CDF)')
plt.legend(fontsize='xx-large')
plt.title('Normal Dağılım')
plt.show()
```

### Uniform Dağılım


```python
# Uniform dağılım (0 ve 1 arasında)
uniform_dist = np.random.random(1000)
uniform_df = pd.DataFrame({'deger': uniform_dist})
uniform_dist = pd.Series(uniform_dist)
```

### Poisson Dağılımı
#### PMF:


```python
from scipy.special import factorial
```

```python
Lambda = 5
x = np.arange(0, 20, 0.1)
y = np.exp(-Lambda)*np.power(Lambda, x) /factorial(x)
plt.figure(figsize=(15,8))
plt.plot(x, y, 'bs')
plt.title('Poisson Dağılımı - Lambda=5')
plt.show()
```

```python
from scipy.stats import poisson
```

```python
Lambda = 5
x = np.arange(0, 20, 0.1)
plt.figure(figsize=(16,5))
plt.plot(x, poisson.cdf(x, mu=Lambda), label='CDF')
plt.plot(x, poisson.pmf(x, mu=Lambda), label='PMF')
plt.title('Poisson CDF')
plt.grid()
```

### Log-Normal Dağılımı


```python
std = 1
mean = 5
x = np.linspace(0,15,200)
dist = lognorm(std, loc=mean)

plt.figure(figsize=(15,6))
plt.plot(x,dist.pdf(x), label='log-normal PDF')
plt.plot(x,dist.cdf(x), label='log-normal CDF')
plt.legend()
plt.title('Log normal PDF')
plt.show()
```
